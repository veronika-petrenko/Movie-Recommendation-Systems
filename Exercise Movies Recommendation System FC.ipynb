```python
from collections import Counter
```

```python
dic = dict(Counter(filtrado['userId'])) #we need to find the users that saw the same 5 movies at the same time, to make correlations calculation possible 
```

```python
dic
```

```python
lista = [] #only choose the ones where the user id repeats at least 5 times 
for i in dic.items(): 
    if i[1] == 5:
        lista.append(i[0])
```

```python
for i in dic.items():
    print(i)
    break
```

```python
lista[0:100] 
```

```python
user100 = filtrado[filtrado['userId'].isin(lista[0:100])] #create df with only 100 users we chose
user100 = user100.reset_index(drop = True)
user100 
```

```python
df2
```

```python
#next step is to find correlations with our user and similar users, we need to create a new df to make calculations easier 
```

```python
columns = ['userId']

for i in df2['movieId'].values:
    columns.append(i)
    
df_agrupado = pd.DataFrame(columns=columns)
df_agrupado 
```

```python
#new df with user id and movie ids in columns 
for i in range(len(lista[0:100])): 
    registro = []
    registro.append(lista[0:100][i])
    for j in df_agrupado.columns[1:]:
        registro.append(float(user100[(user100["userId"] == lista[0:100][i]) & (user100["movieId"] == j)]["rating"].values) )
    df_agrupado.loc[i] = registro
```

```python
df_agrupado
```

```python
from scipy import stats
```

```python
stats.pearsonr(df_agrupado.iloc[0,1:],df2['movieId'])[0]
```

```python
df_agrupado.iloc[0,1:]
```

```python
#calculate correlations 
corr= []
for i in range(len(df_agrupado)):
    corr.append(stats.pearsonr(df_agrupado.iloc[i,1:],df2['rating'])[0])
    
```

```python
df2
```

```python
df_agrupado['corr'] = corr
```

```python
df_agrupado
```

```python
#next step is to find what other similar users watched and our user didn't in order to recommend it to him 

# filter by the same users we were using before but looking for movies that are different from our user

df_agrupado2 = pd.merge(left = df_agrupado, right = users[(users['userId'].isin(lista[0:100])) & (~users['movieId'].isin([1968,    1,    2,  296, 1274]))], how = 'inner', on = ['userId'])
```

```python
df_agrupado2.iloc[0,6] * df_agrupado2.iloc[0,8]
```

```python
#create new column with the multiplication of movie ratings by user and similarity score of the user and our user (corr)
notas = []
for i in range(len(df_agrupado2)):
    notas.append(df_agrupado2.iloc[i,6] * df_agrupado2.iloc[i,8])
```

```python
notas
```

```python
df_agrupado2['notas'] = notas
```

```python
df_agrupado2 #now we have weighted rating for each movie
```

```python
dffinal = df_agrupado2.groupby('movieId').sum().reset_index().sort_values('movieId') #sum the movie ratings 
dffinal
```

```python
dffinal.drop( [1968,1,2,296,1274], axis=1, inplace= True)
```

```python
dffinal
```

```python
#final movie rating is the sum of ratings of each movie divided by the sum of correlations of each user with our user 
ratingfinal = []
for i in range(len(dffinal)):
    if dffinal.iloc[i,2] == 0:
        ratingfinal.append(0)
    else:
        ratingfinal.append(dffinal.iloc[i,4] / dffinal.iloc[i,2])
```

```python
ratingfinal
```

```python
dffinal['ratingfinal'] = ratingfinal
```

```python
dffinal.sort_values('ratingfinal', ascending = False)
```
